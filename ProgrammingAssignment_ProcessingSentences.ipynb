## Download several electronic books from Project Gutenberg. Write a program to scan these texts for any extremely long ##sentences. What is the longest sentence you can find? What syntactic construction(s) are responsible for such long sentences?

```python
#import library and download three texts
import nltk
import numpy as np
test_corpus = nltk.corpus.gutenberg.sents(['austen-emma.txt', 'carroll-alice.txt', 'chesterton-ball.txt'])
```

```python
#test corpus length
len(test_corpus) #323533 words
test_corpus[:20] #look at first 10 sentences
```

```python
#converted to dict. Should make it easier to keep track of sentence lengths
def Convert(lst): 
    res_dct = {lst[i]: lst[i] for i in range(0, len(lst), 1)} 
    return res_dct 

test_corpus_dict = Convert(test_corpus)
```

## Create list of ordered sentence lengths

```python
sentence_lengths = []
num_list = len(test_corpus)
for i in range(num_list):
    #i = 1
    sentencesize = len(test_corpus[i])
    #print(sentencesize)
    if sentencesize < 1:
        sentence_lengths.append(0)
    else:
        sentence_lengths.append(sentencesize)
```

## Find the max

```python
max = 1
for i in range(len(sentence_lengths)):
    if sentence_lengths[i] > max:
        max = sentence_lengths[i]
        maxIndex = i
print(maxIndex) #7045 is index
print(max) #274 is the sentence word count
```
